```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/gdrive/MyDrive/GDP-2006-2022(1).csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(13,7))
df2 = df.drop(['DATE'], axis = 1, inplace = False)
corre = df2.corr()
sns.heatmap(corre, cmap = 'RdBu_r', annot = True, square = True)
```

The graph above shows the correlation matrix table for all the quantitative variables in the csv. We should be careful about predictors with high correlation coefficients to prevent multicollinearity. In this case, GDP PER CAPITA and QUARTERLY REAL GDP are highly correlated, and MED HOUSEHOLD INCOME and CPI are highly correlated.
